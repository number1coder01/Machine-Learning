```python
import numpy as np
a = np.array([1,2,3])
print(type(a)) #vector
```

```python
b = np.array([[1,2,3] , [4,5,6]]) #2D matrix , ||ly tensor and so on
print(b)
```

```python
# dtype
c = np.array([1,2,3] , dtype = float)
d = np.array([1,2,3] , dtype = bool)
e = np.array([1,2,3] , dtype = complex)
print(c)
print(d)
print(e)
```

```python
# np.arange
np.arange(1 , 11 , 2)
```

```python
# with reshape
np.arange(1,11).reshape(5,2)
```

```python
# For initializing arrays

np.ones((3,4))
np.zeros((3,4))
np.random.random((3,4)) # by default dtype = float
```
