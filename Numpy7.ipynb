```python
#Broadcasting ->
  # This described how numpy treats arrays with diff shapes during arithmetic oprns
  # the smaller array is "broadcast" across the larger array so that they have compatible shapes
import numpy as np

# same shape
a = np.arange(6).reshape(2 , 3)
b = np.arange(6 , 12).reshape(2 , 3)
print(a)
print(b)
print(a + b) # Possible
```

```python
# diff shape
a = np.arange(6).reshape(2 , 3)
b = np.arange(3).reshape(1 , 3)
print(a)
print(b)
print(a + b) # Still Possible bcz -> Voh copy kardega same row to all rest to make shape equal
# Check Notes for better visualizn
# Imp Rules for Broadcasting -> 1)Make the 2 arrays have the same number of dimensions
# 2)Make each dimension of the 2 arrays the same size
```
