```python
# Changing Datatypes & Array Operations & Array Functions
```

```python
import numpy as np
a1 = np.arange(10)
a2 = np.arange(12 , dtype = float).reshape(3,4)
a3 = np.arange(40).reshape(2,4,5)
a4 = np.arange(10 , dtype = np.int32)
```

```python
# Change datatype to any other
a1.astype(np.int32) # extra memory usage prevent karne ke liye
a1.astype(bool)
```

```python
# Array Operations
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12 , 24).reshape(3,4)
```

```python
# Scalar Operations (1 scalar with 1 tensor)
print(a1 * 2) #Arithmetic
print(a2 > 15) #Relational
```

```python
# Vector Operations (Dono same shape ki tensors hain)
print(a1 + a2) #Arithmetic
print(a2 > a1) #Relational
```

```python
# Array Functions
a1 = np.random.random((3 , 3))
a1 = np.round(a1 * 100)
print(a1)
```

```python
# Basic Operns
print(np.max(a1))
print(np.min(a1))
print(np.sum(a1))
print(np.prod(a1))

# Har row mei min item -> (axis = 1 => row & 0 => col)#IMP
print(np.min(a1 , axis = 1))
```

```python
# Statstical Oprns
print(np.mean(a1))
print(np.median(a1 , axis = 0))
print(np.std(a1))
print(np.var(a1 , axis = 1))
```

```python
# Trignometric functions
np.sin(a1)
```

```python
# Dot Product (matrix multiplcn)
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12 , 24).reshape(4 , 3)
np.dot(a1 , a2)
```

```python
# Some more functions
a1 = np.arange(1 , 11)
np.log(a1)
np.exp(a1)
a1 = np.random.random((2 , 3))*100
print("normal => \n" , a1)
print("rounded => \n" , np.round(a1))
np.floor(a1)
np.ceil(a1)
```
