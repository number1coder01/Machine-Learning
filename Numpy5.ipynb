```python
# Iterating & Reshaping + Stacking & Spliting
import numpy as np
```

```python
a1 = np.arange(10)
a2 = np.arange(12).reshape(3 , 4)
a3 = np.arange(27).reshape(3 , 3 , 3)

print(a1)
print(a2)
print(a3)
```

```python
for i in a1 :
  print(i)
```

```python
for i in a3 :
  print(a3)
```

```python
for i in np.nditer(a3) :  # nd iter will iterate over all items by 1) Converting 3D to 1D then iterating over that 1D array
  print(i)
```

```python
# Reshaping ->

#Transpose
np.transpose(a2)
a2.T #Same as up

#Ravel -> Kitne bhi dim ki array ko 1D mei convert kardega
np.ravel(a3)
```

```python
# Stacking -> (Shape same hona chahiye isme)
a4 = np.arange(12).reshape(3 , 4)
a5 = np.arange(12 , 24).reshape(3 , 4)

# Vertical Stacking => Ek ke upar ek => 2*2 ke 2 Array then 4*2 ban jayega
print(np.vstack((a4 , a5 , a4 , a5)))

#Horizontal Stacking
print(np.hstack((a4 , a5)))
```

```python
# Spliting -> Opposite of Stacking

# Horizontal Split mei | = line of split #IMP
print(np.hsplit(a4 , 4)) # ENSURE -> Equal size ke rahenge saare
# Vertical Split mei - = line of split #IMP
print(np.vsplit(a5 , 3))
```
