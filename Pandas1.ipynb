```python
import numpy as np
import pandas as pd
```

```python
# Series creation from Lists ->
```

```python
# String
country = ['India' , 'US' , 'Iran' , 'Israel']
pd.Series(country) #yaha pandas => strings ko obj kehta hai
```

```python
# Integer
runs = [13 , 24 ,56 , 78 , 100]
pd.Series(runs)
```

```python
# Custom Index
marks = [100 , 99 , 89 , 78 , 96]
subject = ['maths' , 'chemistry' , 'emotional intelligence' , 'english' , 'physics']

pd.Series(marks , index = subject)
```

```python
# Setting a name for the Series
marks = pd.Series(marks , index = subject , name = 'Marks') # isse aap ek Series class ka object bana rahe
marks
```

```python
# Series from dict

marks = {'maths' : 100 ,
         'chemistry' : 99 ,
         'emotional intelligence' : 89 ,
         'english' : 78 , 'physics' : 97}

marks = pd.Series(marks , name = 'Marks')
```
