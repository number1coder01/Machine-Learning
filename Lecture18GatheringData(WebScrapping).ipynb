```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
}

url = "http://books.toscrape.com/"

response = requests.get(url, headers=headers)

print(response.status_code)   # MUST be 200
print(response.text[:500])    # check what you got
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.find_all('h1')
```

```python
len(soup.find_all('p'))
```

```python
books = soup.find_all("article", class_="product_pod") # find_all tab lagta h jab u think ki iske multiple answers milenge
for book in books:
    title = book.find("h3").find("a")["title"]
    price = book.find("p", class_="price_color").text #specific class nikaalni from 'p'
    rating = book.find("p")["class"][1]   # since class returned as a list -> ["star-rating", "Three"]

    print(title, price, rating , '\n')
```

```python
#Now storing the above in a dataframe ->
data = []
```

```python
base_url = "http://books.toscrape.com/catalogue/page-{}.html"

headers = {
    "User-Agent": "Mozilla/5.0"
}

for page in range(1, 6):  # first 5 pages
    url = base_url.format(page)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    books = soup.find_all("article", class_="product_pod")

    for book in books:
        title = book.find("h3").find("a")["title"]
        price = book.find("p", class_="price_color").text
        rating = book.find("p")["class"][1]

        data.append({
            "Title": title,
            "Price": price,
            "Rating": rating
        })
```

```python
df = pd.DataFrame(data)
```
