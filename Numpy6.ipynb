```python
# TOPIC 1 =>
  #  Numpy Array vs Python Lists
import numpy as np
import time as t
```

```python
# Speed ->

# List
a = [i for i in range(10000000)]
b = [i for i in range(10000000, 20000000)]

c = []
start = t.time()
for i in range(len(a)) :
  c.append(a[i] + b[i])

print(t.time() - start)

# Numpy Array
a = np.arange(10000000)
b = np.arange(10000000 , 20000000)

start = t.time()
c = a + b
print(t.time() - start)
```

```python
# Memory ->

# List
a = [i for i in range(10000000)]
import sys

print(sys.getsizeof(a))
# Numpy Array
a = np.arange(10000000 ,dtype=np.int32) #normal int64 same size occupy karega
print(sys.getsizeof(a))
```

```python
#TOPIC 2 -> Advanced Indexing
a = np.arange(24).reshape(6,4)
# 1)Fancy Indexing -> Jab NO pattern for fetching then give a list and usme joh chahiye voh provide kardo
print(a[[0,2,3,5]])
print(a[:,[0,2,3]]) # All rows but only 0th 2nd and 3rd column
```

```python
a = np.random.randint(1 , 100 , 24).reshape(6 , 4) # 24 items randomly from 1 to 100
print(a)
# 2)Boolean Indexing -> Fetching on the basis of a condn
b = a > 50 # Now this is a boolean Array and can be used as a BOOLEAN MASK #IMP
print(b)
print(a[b]) # Jaha jaha True hogi value only those
print(a[~(a % 2 == 0) & (a > 50)]) # Yaha bitwise AND aayega kyuki boolean values ke saath deal kar rahe
```
