```python
# Indexing & Slicing
import numpy as np
```

```python
a1 = np.arange(10)
a2 = np.arange(12).reshape(3 , 4)
a3 = np.arange(27).reshape(3 , 3 , 3)

print(a1)
print(a2)
print(a3)
```

```python
# Indexing
print(a1[-1])
print(a1[2])
print(a2[0 , 2])
print(a3[0 , 1 , 1]) # same as [0][1][1]
```

```python
# Slicing -> Ek saath multiple items fetch
# Trick -> Ask urself Konsi rows chahiye and konse cols then write them
print(a1[2 : 5]) # sp to ep - 1 , agar ep last idx tak chahiye toh write nothing i.e.(1:)
print(a1[2 : 5 : 2])
```

```python
# 2D Slicing
print(a2[0 , :]) # Pehla row with all columns
print(a2[ : , 2]) # 3rd col with all rows
print(a2[1: , 1 : 3]) # 1st row onwards and col 1 to 2
print(a2[::2 , ::3])
print(a2[0::2 , 1::2])
print(a2[1 , 0 :: 3])
```

```python
# 3D Slicing
print(a3[0 , 1])
print(a3[1 , : , 1]) # 2nd wali 2D array ka 1st col and all rows
print(a3[::2 , 0 , :: 2])
```
