```python
#3. OneHotEncoding using Sklearn -> yeh better rehta hai than pandas ka ohe
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# drop first h isliye 9 ki jagah 7 hi aayenge
ohe = OneHotEncoder(drop='first', sparse_output=False, dtype=np.int32)
```

```python
# yeh ohe ek sparse matrix produce karega jisme no. cols = categories jitni di hai
# but since sparse = false toh yeh o/p will be a numpy array
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
```

```python
X_test_new = ohe.transform(X_test[['fuel','owner']])
```

```python
# yaha 7 hai kyuki isme brand aur km_driven wala feature nahi h
X_train_new.shape
```

```python
# appending the new_X_Train into the brand and km_driven wala array
np.hstack((X_train[['brand','km_driven']].values,X_train_new))
```
